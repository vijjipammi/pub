### Import Standard Pytorch Imports

```python
import torch
import torch.optim as optim
from torch.nn import Linear, MSELoss, functional as F
import numpy as np
```

### Create Data, X and y

```python
X= np.linspace(1,20,1000)
```

```python
y = 3*X*X + 5
```

```python
X.shape
```

```python
y.shape
```

```python
X =X.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
X.shape
```

```python
y.shape
```

### Split Test and Train Data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.2, shuffle=True, random_state=35)
```

```python
X_train.shape
```

```python
X_train=torch.from_numpy(X_train).float()
X_test = torch.from_numpy(X_test).float()
y_train = torch.from_numpy(y_train).float()
y_test = torch.from_numpy(y_test).float()
```

```python
type(X_train)
```

```python
X_train.shape
```

### Define the Model

```python
class Net(torch.nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = torch.nn.Linear(1, 12)
        self.fc2 = torch.nn.Linear(12, 12)
        self.fc3 = torch.nn.Linear(12, 12)
        self.fc4 = torch.nn.Linear(12, 1)

    def forward(self, x):
        x  = self.fc1(F.relu(x))
        x  = self.fc2(F.relu(x))
        x  = self.fc3(F.relu(x))
        x  = self.fc4(x)
        return x

    
    
```

```python
model =Net()
```

```python
for param in model.parameters():
    print(param)
```

```python
for children in model.children():
    print(children)
```

```python
model = Net()

# The nn package also contains definitions of popular loss functions; in this
# case we will use Mean Squared Error (MSE) as our loss function.
loss_fn = torch.nn.MSELoss()
t=0

optimizer = optim.Adam(model.parameters(), lr=0.001)
for epch in range(100):
    for x,y in zip(X_train,y_train):
        y_pred = model(x)
        loss = loss_fn(y_pred, y)
        
        t=t+1
        if t % 100 == 99:
            print(t, loss.item())

        model.zero_grad()
        loss.backward()
        optimizer.step()
```

```python
y_pred=model(X_test)
```
